### BÖLÜM 3: HF Pipeline oluşturalım

---



```python
def tokenize(data):
    results = tokenizer(data["review"],
                        truncation=True,
                        max_length=30,
                        return_overflowing_tokens=False,
                        return_length=True)
    batch = []
    for length, input_id in zip(results["length"], results["input_ids"]):
        if length == 30:
            batch.append(input_id)
    return {"input_ids":batch}
```

```python
tokenized_df = df.map(tokenize, batched=True, remove_columns=df["train"].column_names)
tokenized_df
```

```python
save_data(tokenized_df,"tokenized_df")
```

```python
tokenized_df = load_data("tokenized_df")
```

```python
tokenized_df["train"][41]
```

```python
# HF Data Collator ile veriyi batchler halinde hazırlayalım.
from transformers import DataCollatorForLanguageModeling

tokenizer.pad_token = tokenizer.eos_token
dc = DataCollatorForLanguageModeling(tokenizer, mlm=False, return_tensors="tf")
```

```python
out = dc([tokenized_df["train"][i] for i in range(3)])
for k in out:
    print(f"{k} shape: {out[k].shape}")
```

```python
# HF dataseti TF datasete çevirelim.
tf_train_df = tokenized_df["train"].to_tf_dataset(
    columns=["input_ids","attention_mask","labels"],
    collate_fn=dc,
    shuffle=True,
    batch_size=32
)

tf_eval_df = tokenized_df["validation"].to_tf_dataset(
    columns=["input_ids","attention_mask","labels"],
    collate_fn=dc,
    shuffle=False,
    batch_size=32
)
```

```python
len(tf_train_df)
```

### BÖLÜM 4: Modelimizi eğitelim

---



gpt-2 modelimizi ağırlıkları olmadan sadece model mimarisini HF den indirip ayarlamaları yapalım.Modeli sıfırdan train edeceğiz.


```python
from transformers import AutoTokenizer, TFGPT2LMHeadModel, AutoConfig

conf = AutoConfig.from_pretrained(
    "gpt2",
    vocab_size=len(tokenizer),
    n_ctx=30,
    bos_token_id=tokenizer.bos_token_id,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
conf.eos_token_id = 0
```

```python
conf
```

```python
save_data(conf,"conf")
```

```python
model = TFGPT2LMHeadModel(conf)
model(model.dummy_inputs) # dummy verip modeli build edelim
model.summary()
```

```python
# optimizer oluşturalım.
from transformers import create_optimizer

n_train_steps = len(tf_train_df) # 9811
optimizer, lr_schedule = create_optimizer(
    init_lr=3e-5,
    num_warmup_steps=1000,
    num_train_steps=n_train_steps,
    weight_decay_rate=0.01
)
```

```python
# modeli derleyelim.
model.compile(optimizer=optimizer)

tf.keras.mixed_precision.set_global_policy("mixed_float16") # GPU varsa kullanılacaktır.
```

```python
save_data(model,"model")
```

### BÖLÜM 5: HF Text Generation Pipeline

```python
from transformers import pipeline
from transformers import AutoTokenizer, TFGPT2LMHeadModel, AutoConfig
from datasets import load_dataset

dataset = load_dataset("selcuktekgoz/turkishReviews-230k", split="validation")
model = TFGPT2LMHeadModel.from_pretrained("selcuktekgoz/turkishReviews-230k")
tokenizer = AutoTokenizer.from_pretrained("selcuktekgoz/turkishReviews-230k")

pipe = pipeline("text-generation",model=model, tokenizer= tokenizer, device=0)
```

```python
import gradio as gr
```

```python
examples = ["Casper Nirvana Pişmanlıktır",
            "Karaca Home Parasını Alıp",
            "Samsung televizyonum sürekli"]
```

```python
def generate_text(prompt):
    result = pipe(prompt,num_return_sequences=1)
    return result[0]["generated_text"]
```

```python
iface = gr.Interface(fn=generate_text, inputs="text", outputs="text", examples=examples)
```
